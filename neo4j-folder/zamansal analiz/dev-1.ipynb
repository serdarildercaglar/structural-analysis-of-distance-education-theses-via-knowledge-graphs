```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri
yıllar = [1997, 1999, 2000, 2001, 2002, 2003, 2004, 2006, 2007, 2008, 2009, 
          2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 
          2020, 2021, 2022, 2023, 2024, 2025]
tez_sayıları = [2, 1, 1, 4, 11, 10, 1, 1, 5, 6, 7, 
                17, 14, 13, 9, 24, 19, 16, 24, 19, 30, 
                24, 152, 194, 87, 10, 2]

# Veriyi DataFrame'e dönüştürme
df = pd.DataFrame({'Yıl': yıllar, 'Tez Sayısı': tez_sayıları})

# Grafik ayarları
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Pandemi öncesi ve sonrası için renk ayarları
colors = ['#1f77b4' if year < 2020 else '#d62728' for year in yıllar]

# Çubuk grafiği oluşturma
ax = sns.barplot(x='Yıl', y='Tez Sayısı', data=df, palette=colors)

# Pandemi başlangıcını gösteren dikey çizgi
plt.axvline(x=19.5, color='gray', linestyle='--', alpha=0.7)
plt.text(19.7, max(tez_sayıları)*0.9, 'COVID-19 Pandemi Başlangıcı', fontsize=12, color='gray')

# Başlık ve etiketler
plt.title('Şekil 12: Yıllara Göre Uzaktan Eğitim Tezlerinin Dağılımı (1997-2025)', fontsize=16, pad=20)
plt.xlabel('Yıl', fontsize=14)
plt.ylabel('Tez Sayısı', fontsize=14)

# X ekseni düzenlemesi
plt.xticks(rotation=45)

# Değerlerin gösterilmesi
for i, v in enumerate(tez_sayıları):
    if v > 30:  # Sadece belirli bir değerden büyük olanları göster
        ax.text(i, v + 5, str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('sekil12_yillara_gore_tez_sayilari.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Anahtar kavramlar ve yıllara göre değerleri
kavramlar = [
    "distance education",
    "student motivation",
    "student motivation problem",
    "attitude towards distance education",
    "hybrid education model",
    "technology integration in education"
]

# Kavramların yıllara göre sayıları (sadece önemli yıllar gösteriliyor)
yillar = [2010, 2015, 2019, 2020, 2021, 2022, 2023, 2024]

# Kavramların yıllara göre sayısı (verilerden çıkarılmış)
data = {
    "distance education": [7, 10, 18, 20, 143, 183, 99, 10],
    "student motivation": [0, 0, 2, 0, 9, 20, 12, 0],
    "student motivation problem": [0, 0, 1, 2, 8, 16, 15, 2],
    "attitude towards distance education": [1, 1, 2, 5, 22, 29, 0, 0],
    "hybrid education model": [0, 0, 0, 0, 1, 2, 4, 2],
    "technology integration in education": [3, 5, 1, 2, 12, 16, 5, 2]
}

# DataFrame oluşturma
df_list = []
for kavram in kavramlar:
    for i, yil in enumerate(yillar):
        df_list.append({
            "Yıl": yil,
            "Kavram": kavram,
            "Sayı": data[kavram][i]
        })

df = pd.DataFrame(df_list)

# Grafik ayarları
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Çizgi grafiği oluşturma
ax = sns.lineplot(
    data=df,
    x="Yıl",
    y="Sayı",
    hue="Kavram",
    style="Kavram",
    markers=True,
    dashes=False,
    markersize=10,
    linewidth=2.5
)

# Pandemi başlangıcını gösteren dikey çizgi
plt.axvline(x=2019.5, color='gray', linestyle='--', alpha=0.7)
plt.text(2019.7, max(df['Sayı'])*0.9, 'COVID-19 Pandemi Başlangıcı', fontsize=12, color='gray')

# Başlık ve etiketler
plt.title('Şekil 14: Anahtar Kavramların Yıllara Göre Gelişimi (2010-2024)', fontsize=16, pad=20)
plt.xlabel('Yıl', fontsize=14)
plt.ylabel('Kullanım Sayısı', fontsize=14)

# Y ekseni sınırlaması
plt.ylim(0, 190)

# Lejantı düzenleme
plt.legend(title='Kavram', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('sekil14_anahtar_kavramlarin_gelisimi.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Ana ilişki türleri
iliskiler = [
    "EXAMINES",
    "REQUIRES",
    "FACES",
    "USES",
    "ENHANCES",
    "HINDERS",
    "ADDRESSES"
]

# Önemli yıllar
yillar = [2010, 2015, 2019, 2020, 2021, 2022, 2023, 2024]

# İlişkilerin yıllara göre sayısı (verilerden normalized)
# Gerçek sayılar çok büyük olduğu için grafik okunabilirliği için normalize edilmiştir
data = {
    "EXAMINES": [6632, 7410, 11700, 9364, 59290, 75676, 33941, 3900],
    "REQUIRES": [2681, 3825, 6867, 7498, 53312, 68010, 36701, 3708],
    "FACES": [3265, 3436, 3132, 5300, 29823, 40634, 24026, 2810],
    "USES": [2309, 2275, 2172, 3776, 20497, 27983, 16779, 1982],
    "ENHANCES": [856, 1170, 1823, 1947, 13792, 17515, 9210, 970],
    "HINDERS": [607, 856, 1577, 1721, 12376, 15985, 8637, 870],
    "ADDRESSES": [235, 367, 495, 444, 3207, 4268, 2186, 212]
}

# Log ölçeğinde görselleştirme için veriyi hazırlama
df_list = []
for iliski in iliskiler:
    for i, yil in enumerate(yillar):
        df_list.append({
            "Yıl": yil,
            "İlişki Türü": iliski,
            "Sayı": data[iliski][i],
            "Log Sayı": np.log10(data[iliski][i]) if data[iliski][i] > 0 else 0
        })

df = pd.DataFrame(df_list)

# İlişki türleri için renkler ve etiketler
colors = {
    'EXAMINES': '#1f77b4',
    'REQUIRES': '#ff7f0e',
    'FACES': '#2ca02c',
    'USES': '#d62728',
    'ENHANCES': '#9467bd',
    'HINDERS': '#8c564b',
    'ADDRESSES': '#e377c2'
}

labels = {
    'EXAMINES': 'İnceler',
    'REQUIRES': 'Gerektirir',
    'FACES': 'Karşılaşır',
    'USES': 'Kullanır',
    'ENHANCES': 'Geliştirir',
    'HINDERS': 'Engeller',
    'ADDRESSES': 'Ele Alır'
}

# İlişki türlerini yeniden etiketleme
df['İlişki Türü Etiketi'] = df['İlişki Türü'].map(labels)

# Grafik ayarları
plt.figure(figsize=(14, 10))
sns.set_style("whitegrid")

# Log ölçekli çizgi grafiği oluşturma
ax = sns.lineplot(
    data=df,
    x="Yıl",
    y="Log Sayı",
    hue="İlişki Türü Etiketi",
    style="İlişki Türü Etiketi",
    markers=True,
    dashes=False,
    markersize=10,
    linewidth=2.5,
    palette=colors.values()
)

# Pandemi başlangıcını gösteren dikey çizgi
plt.axvline(x=2019.5, color='gray', linestyle='--', alpha=0.7)
plt.text(2019.7, 4.8, 'COVID-19 Pandemi Başlangıcı', fontsize=12, color='gray')

# Başlık ve etiketler
plt.title('Şekil 15: İlişki Türlerinin Zamansal Gelişimi (2010-2024, Logaritmik Ölçek)', fontsize=16, pad=20)
plt.xlabel('Yıl', fontsize=14)
plt.ylabel('İlişki Sayısı (log10 ölçeği)', fontsize=14)

# Y ekseni düzenlemesi
plt.yticks([1, 2, 3, 4, 5], ['10', '100', '1,000', '10,000', '100,000'])

# Lejantı düzenleme
plt.legend(title='İlişki Türü', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('sekil15_iliski_turlerinin_zamansal_gelisimi.png', dpi=300)
plt.show()
```
