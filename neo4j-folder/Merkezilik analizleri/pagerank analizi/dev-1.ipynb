```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Tema ayarları
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Hardcode edilmiş veriler
# Genel PageRank top 20
general_pagerank_data = [
    {"name": "distance education", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 18.596022364742225},
    {"name": "distance education", "entity_type": "SOLUTION_APPROACH", "pagerank_score": 18.596022364742225},
    {"name": "student autonomy", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 9.87179207861687},
    {"name": "academic achievement", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 5.748961311144132},
    {"name": "academic achievement", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 5.748961311144132},
    {"name": "student motivation", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 5.4590624111617965},
    {"name": "student motivation", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 5.4590624111617965},
    {"name": "student achievement", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 4.057241611287906},
    {"name": "student achievement", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 4.057241611287906},
    {"name": "student motivation problem", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 3.622868817286202},
    {"name": "transactional distance", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 3.4590020439076574},
    {"name": "effectiveness of online education", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 3.4020359758924283},
    {"name": "student participation", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 3.3025314569337856},
    {"name": "attitude towards distance education", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 3.2721578879933784},
    {"name": "attitude towards distance education", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 3.2721578879933784},
    {"name": "mathematics education", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 3.0956131876271598},
    {"name": "student satisfaction", "entity_type": "PROBLEM_CHALLENGE", "pagerank_score": 3.0741715260950344},
    {"name": "student satisfaction", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 3.0741715260950344},
    {"name": "technology integration in education", "entity_type": "FOCUS_AREA_THEME", "pagerank_score": 2.978062476292467},
    {"name": "technology integration in education", "entity_type": "SOLUTION_APPROACH", "pagerank_score": 2.9780624762924663}
]

# Varlık türlerine göre PageRank istatistikleri
entity_stats = [
    {"entity_type": "FOCUS_AREA_THEME", "node_count": 469, "avg_pagerank": 0.5136986164372358, 
     "min_pagerank": 0.15, "max_pagerank": 18.596022364742225, "stdev_pagerank": 1.1385719051732524},
    {"entity_type": "PROBLEM_CHALLENGE", "node_count": 850, "avg_pagerank": 0.3328167162897541,
     "min_pagerank": 0.15, "max_pagerank": 5.748961311144131, "stdev_pagerank": 0.4377708769515447},
    {"entity_type": "SOLUTION_APPROACH", "node_count": 1407, "avg_pagerank": 0.2155586633682448,
     "min_pagerank": 0.15, "max_pagerank": 18.596022364742225, "stdev_pagerank": 0.5219057171089588},
    {"entity_type": "STAKEHOLDER", "node_count": 414, "avg_pagerank": 0.17164247747042533,
     "min_pagerank": 0.15, "max_pagerank": 1.1720061597922473, "stdev_pagerank": 0.09080486666384323}
]

# Dört merkezilik türünün karşılaştırması
centrality_comparison = [
    {"name": "distance education", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 572.0,
     "betweenness_centrality": 697732.630625, "closeness_centrality": 0.38186923385479343, "pagerank_score": 18.596022364742225},
    {"name": "distance education", "entity_type": "SOLUTION_APPROACH", "degree_centrality": 572.0,
     "betweenness_centrality": 697732.630625, "closeness_centrality": 0.38186923385479343, "pagerank_score": 18.596022364742225},
    {"name": "student autonomy", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 3.0,
     "betweenness_centrality": 46408.495057319065, "closeness_centrality": 0.30454254638515676, "pagerank_score": 9.87179207861687},
    {"name": "academic achievement", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 5.0,
     "betweenness_centrality": 61696.13620195766, "closeness_centrality": 0.34643377001455605, "pagerank_score": 5.748961311144131},
    {"name": "academic achievement", "entity_type": "PROBLEM_CHALLENGE", "degree_centrality": 5.0,
     "betweenness_centrality": 61696.13620195766, "closeness_centrality": 0.34643377001455605, "pagerank_score": 5.748961311144131},
    {"name": "student motivation", "entity_type": "PROBLEM_CHALLENGE", "degree_centrality": 1.0,
     "betweenness_centrality": 20408.014442940268, "closeness_centrality": 0.3174391463821274, "pagerank_score": 5.4590624111617965},
    {"name": "student motivation", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 1.0,
     "betweenness_centrality": 20408.014442940268, "closeness_centrality": 0.3174391463821274, "pagerank_score": 5.4590624111617965},
    {"name": "student achievement", "entity_type": "PROBLEM_CHALLENGE", "degree_centrality": 0.0,
     "betweenness_centrality": 0.0, "closeness_centrality": 0.30855365474339036, "pagerank_score": 4.057241611287906},
    {"name": "student achievement", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 0.0,
     "betweenness_centrality": 0.0, "closeness_centrality": 0.30855365474339036, "pagerank_score": 4.057241611287906},
    {"name": "student motivation problem", "entity_type": "PROBLEM_CHALLENGE", "degree_centrality": 2.0,
     "betweenness_centrality": 192.76963626030127, "closeness_centrality": 0.3031055900621118, "pagerank_score": 3.622868817286203},
    {"name": "transactional distance", "entity_type": "PROBLEM_CHALLENGE", "degree_centrality": 3.0,
     "betweenness_centrality": 56705.580041306865, "closeness_centrality": 0.24180848361696725, "pagerank_score": 3.459002043907658},
    {"name": "effectiveness of online education", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 4.0,
     "betweenness_centrality": 96013.95453160214, "closeness_centrality": 0.3019346653980336, "pagerank_score": 3.4020359758924283},
    {"name": "student participation", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 2.0,
     "betweenness_centrality": 21132.118880396647, "closeness_centrality": 0.30493273542600896, "pagerank_score": 3.302531456933785},
    {"name": "attitude towards distance education", "entity_type": "FOCUS_AREA_THEME", "degree_centrality": 13.0,
     "betweenness_centrality": 44229.34449449078, "closeness_centrality": 0.2970358814352574, "pagerank_score": 3.2721578879933784},
    {"name": "attitude towards distance education", "entity_type": "PROBLEM_CHALLENGE", "degree_centrality": 13.0,
     "betweenness_centrality": 44229.34449449078, "closeness_centrality": 0.2970358814352574, "pagerank_score": 3.2721578879933784}
]

# Renk paleti tanımlaması
color_map = {
    'STAKEHOLDER': '#FF6B6B',           # Kırmızı
    'PROBLEM_CHALLENGE': '#4ECDC4',     # Teal
    'SOLUTION_APPROACH': '#45B7D1',     # Mavi
    'FOCUS_AREA_THEME': '#96CEB4'       # Yeşil
}

# 1. Şekil 5.2.9: En Yüksek PageRank Skoruna Sahip Varlıklar
def create_top_pagerank_chart():
    df = pd.DataFrame(general_pagerank_data)
    df['label'] = df['name'] + ' (' + df['entity_type'] + ')'
    
    plt.figure(figsize=(14, 10))
    colors = [color_map[entity_type] for entity_type in df['entity_type']]
    
    bars = plt.barh(range(len(df)), df['pagerank_score'], color=colors)
    
    # Y ekseni etiketleri
    plt.yticks(range(len(df)), [label if len(label) <= 50 else label[:47] + '...' 
                                for label in df['label']], fontsize=9)
    
    # Ekseni formatı
    plt.xlabel('PageRank Skoru', fontsize=12, fontweight='bold')
    plt.ylabel('Varlıklar', fontsize=12, fontweight='bold')
    plt.title('En Yüksek PageRank Skoruna Sahip Varlıklar (Top 20)', 
              fontsize=14, fontweight='bold', pad=20)
    
    # Grid
    plt.grid(axis='x', alpha=0.3, linestyle='--')
    
    # Değer etiketleri
    for i, (bar, score) in enumerate(zip(bars, df['pagerank_score'])):
        plt.text(score + 0.2, bar.get_y() + bar.get_height()/2, 
                f'{score:.2f}', va='center', fontsize=8)
    
    # Legend
    legend_elements = [plt.Rectangle((0,0),1,1, color=color_map[et], label=et.replace('_', ' ').title())
                      for et in color_map.keys()]
    plt.legend(handles=legend_elements, loc='lower right', fontsize=10)
    
    plt.tight_layout()
    plt.savefig('Şekil_5_2_9_En_Yüksek_PageRank_Skoruna_Sahip_Varlıklar.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 2. Şekil 5.2.10: Varlık Türlerine Göre PageRank Dağılımı
def create_pagerank_distribution():
    # Simülasyon: İstatistiklere dayalı veri oluşturma
    np.random.seed(42)
    
    data_for_boxplot = []
    labels = []
    colors_for_box = []
    
    for stat in entity_stats:
        entity_type = stat['entity_type']
        mean = stat['avg_pagerank']
        std = stat['stdev_pagerank']
        count = min(stat['node_count'], 1000)  # Performans için limit
        
        # Log-normal dağılım kullanarak gerçekçi veri oluştur
        data = np.random.lognormal(mean=np.log(mean), sigma=std/mean, size=count)
        # Min ve max değerleri ile sınırla
        data = np.clip(data, stat['min_pagerank'], stat['max_pagerank'])
        
        data_for_boxplot.append(data)
        labels.append(entity_type.replace('_', ' ').title())
        colors_for_box.append(color_map[entity_type])
    
    plt.figure(figsize=(12, 8))
    
    # Box plot oluştur
    box_plot = plt.boxplot(data_for_boxplot, patch_artist=True, labels=labels)
    
    # Renkleri uygula
    for patch, color in zip(box_plot['boxes'], colors_for_box):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    plt.ylabel('PageRank Skoru', fontsize=12, fontweight='bold')
    plt.xlabel('Varlık Türleri', fontsize=12, fontweight='bold')
    plt.title('Varlık Türlerine Göre PageRank Dağılımı', 
              fontsize=14, fontweight='bold', pad=20)
    
    # Y eksenini log scale yap
    plt.yscale('log')
    plt.grid(axis='y', alpha=0.3, linestyle='--')
    
    # X ekseni etiketlerini döndür
    plt.xticks(rotation=45, ha='right')
    
    plt.tight_layout()
    plt.savefig('Şekil_5_2_10_Varlık_Türlerine_Göre_PageRank_Dağılımı.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 3. Şekil 5.2.11: Dört Merkezilik Türünün Karşılaştırması
def create_centrality_comparison():
    df = pd.DataFrame(centrality_comparison)
    df['label'] = df['name'] + ' (' + df['entity_type'] + ')'
    
    # Normalizasyon (0-1 arası)
    metrics = ['degree_centrality', 'betweenness_centrality', 'closeness_centrality', 'pagerank_score']
    df_norm = df.copy()
    
    for metric in metrics:
        max_val = df[metric].max()
        min_val = df[metric].min()
        df_norm[metric + '_norm'] = (df[metric] - min_val) / (max_val - min_val)
    
    # İlk 10 varlık için görselleştirme
    df_plot = df_norm.head(10)
    
    fig, ax = plt.subplots(figsize=(14, 10))
    
    # Bar genişlikleri
    bar_width = 0.2
    x = np.arange(len(df_plot))
    
    # Her metrik için barlar
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
    metric_names = ['Derece Merkeziliği', 'Arasındalık Merkeziliği', 
                   'Yakınlık Merkeziliği', 'PageRank']
    
    for i, (metric, name, color) in enumerate(zip(
        ['degree_centrality_norm', 'betweenness_centrality_norm', 
         'closeness_centrality_norm', 'pagerank_score_norm'],
        metric_names, colors)):
        
        offset = (i - 1.5) * bar_width
        ax.bar(x + offset, df_plot[metric], bar_width, 
               label=name, color=color, alpha=0.8)
    
    # Eksen formatı
    ax.set_xlabel('Varlıklar', fontsize=12, fontweight='bold')
    ax.set_ylabel('Normalize Edilmiş Merkezilik Skoru', fontsize=12, fontweight='bold')
    ax.set_title('Dört Merkezilik Türünün Karşılaştırması (Top 10)', 
                 fontsize=14, fontweight='bold', pad=20)
    
    # X ekseni etiketleri
    labels = [label if len(label) <= 40 else label[:37] + '...' 
              for label in df_plot['label']]
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=9)
    
    # Legend ve grid
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    plt.tight_layout()
    plt.savefig('Şekil_5_2_11_Dört_Merkezilik_Türünün_Karşılaştırması.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# Tüm grafikları oluştur
if __name__ == "__main__":
    print("PageRank analizi görselleştirmeleri oluşturuluyor...")
    
    create_top_pagerank_chart()
    print("✓ Şekil 5.2.9 oluşturuldu")
    
    create_pagerank_distribution()
    print("✓ Şekil 5.2.10 oluşturuldu")
    
    create_centrality_comparison()
    print("✓ Şekil 5.2.11 oluşturuldu")
    
    print("Tüm görselleştirmeler tamamlandı!")
```
