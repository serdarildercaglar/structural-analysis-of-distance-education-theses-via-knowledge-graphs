```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Font settings
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 11

# Entity type degree centrality statistics
stats_data = [
    {"entity_type": "STAKEHOLDER", "node_count": 414, "avg_degree": 8.47, "min_degree": 0.0, "max_degree": 487.0, "stdev_degree": 37.09},
    {"entity_type": "FOCUS_AREA_THEME", "node_count": 469, "avg_degree": 3.16, "min_degree": 0.0, "max_degree": 572.0, "stdev_degree": 26.90},
    {"entity_type": "SOLUTION_APPROACH", "node_count": 1407, "avg_degree": 2.60, "min_degree": 0.0, "max_degree": 572.0, "stdev_degree": 15.62},
    {"entity_type": "PROBLEM_CHALLENGE", "node_count": 850, "avg_degree": 0.31, "min_degree": 0.0, "max_degree": 13.0, "stdev_degree": 0.96}
]

# Create DataFrame
df_stats = pd.DataFrame(stats_data)

# Color palette
colors = ['#2E86AB', '#F18F01', '#A23B72', '#C73E1D']

# Create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Average degree centrality
bars1 = ax1.bar(df_stats['entity_type'], df_stats['avg_degree'], 
                color=colors, alpha=0.8)
ax1.set_title('Average Degree Centrality by Entity Type', fontsize=13, fontweight='bold')
ax1.set_ylabel('Average Degree', fontsize=11)
ax1.grid(axis='y', alpha=0.3)

# Add values on top of bars
for bar, value in zip(bars1, df_stats['avg_degree']):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{value:.2f}',
             ha='center', va='bottom', fontweight='bold', fontsize=10)

# 2. Maximum degree centrality
bars2 = ax2.bar(df_stats['entity_type'], df_stats['max_degree'], 
                color=colors, alpha=0.8)
ax2.set_title('Maximum Degree Centrality by Entity Type', fontsize=13, fontweight='bold')
ax2.set_ylabel('Maximum Degree', fontsize=11)
ax2.grid(axis='y', alpha=0.3)

# Add values on top of bars
for bar, value in zip(bars2, df_stats['max_degree']):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 5,
             f'{int(value)}',
             ha='center', va='bottom', fontweight='bold', fontsize=10)

# 3. Standard deviation
bars3 = ax3.bar(df_stats['entity_type'], df_stats['stdev_degree'], 
                color=colors, alpha=0.8)
ax3.set_title('Degree Centrality Standard Deviation by Entity Type', fontsize=13, fontweight='bold')
ax3.set_ylabel('Standard Deviation', fontsize=11)
ax3.grid(axis='y', alpha=0.3)

# Add values on top of bars
for bar, value in zip(bars3, df_stats['stdev_degree']):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{value:.2f}',
             ha='center', va='bottom', fontweight='bold', fontsize=10)

# 4. Entity count
bars4 = ax4.bar(df_stats['entity_type'], df_stats['node_count'], 
                color=colors, alpha=0.8)
ax4.set_title('Entity Count by Type', fontsize=13, fontweight='bold')
ax4.set_ylabel('Entity Count', fontsize=11)
ax4.grid(axis='y', alpha=0.3)

# Add values on top of bars
for bar, value in zip(bars4, df_stats['node_count']):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 20,
             f'{int(value)}',
             ha='center', va='bottom', fontweight='bold', fontsize=10)

# Rotate x-axis labels
for ax in [ax1, ax2, ax3, ax4]:
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
#savefig
plt.savefig('5.2.2_degree_centrality_statistics.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()

# Print summary table
print("=== DEGREE CENTRALITY STATISTICS BY ENTITY TYPE ===")
print(f"{'Entity Type':<20} {'Count':<8} {'Avg':<8} {'Min':<8} {'Max':<8} {'Std':<8}")
print("-" * 65)
for _, row in df_stats.iterrows():
    print(f"{row['entity_type']:<20} {row['node_count']:<8} "
          f"{row['avg_degree']:<8.2f} {row['min_degree']:<8.1f} "
          f"{row['max_degree']:<8.1f} {row['stdev_degree']:<8.2f}")
```
