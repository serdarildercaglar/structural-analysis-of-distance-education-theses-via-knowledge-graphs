```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Tema ayarları
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Renk paleti tanımlaması
color_map = {
    'STAKEHOLDER': '#FF6B6B',           # Kırmızı
    'PROBLEM_CHALLENGE': '#4ECDC4',     # Teal
    'SOLUTION_APPROACH': '#45B7D1',     # Mavi
    'FOCUS_AREA_THEME': '#96CEB4'       # Yeşil
}

# Hardcode edilmiş veriler

# 1. Topluluk boyut dağılımı
community_size_data = [
    {"community_size": 329, "number_of_communities": 1, "sample_community_ids": [493]},
    {"community_size": 296, "number_of_communities": 1, "sample_community_ids": [1491]},
    {"community_size": 228, "number_of_communities": 1, "sample_community_ids": [1541]},
    {"community_size": 153, "number_of_communities": 1, "sample_community_ids": [2345]},
    {"community_size": 119, "number_of_communities": 1, "sample_community_ids": [199]},
    {"community_size": 111, "number_of_communities": 1, "sample_community_ids": [2257]},
    {"community_size": 63, "number_of_communities": 1, "sample_community_ids": [418]},
    {"community_size": 49, "number_of_communities": 1, "sample_community_ids": [1424]},
    {"community_size": 48, "number_of_communities": 2, "sample_community_ids": [1281, 73]},
    {"community_size": 39, "number_of_communities": 1, "sample_community_ids": [2727]},
    {"community_size": 36, "number_of_communities": 1, "sample_community_ids": [640]},
    {"community_size": 33, "number_of_communities": 1, "sample_community_ids": [786]},
    {"community_size": 24, "number_of_communities": 1, "sample_community_ids": [852]},
    {"community_size": 20, "number_of_communities": 1, "sample_community_ids": [1967]},
    {"community_size": 19, "number_of_communities": 2, "sample_community_ids": [2574, 2828]},
    {"community_size": 16, "number_of_communities": 1, "sample_community_ids": [1699]},
    {"community_size": 15, "number_of_communities": 1, "sample_community_ids": [236]},
    {"community_size": 11, "number_of_communities": 1, "sample_community_ids": [974]},
    {"community_size": 10, "number_of_communities": 4, "sample_community_ids": [735, 765, 2066, 1863]},
    {"community_size": 9, "number_of_communities": 2, "sample_community_ids": [1012, 1685]},
    {"community_size": 8, "number_of_communities": 3, "sample_community_ids": [310, 2724, 1586]},
    {"community_size": 7, "number_of_communities": 5, "sample_community_ids": [862, 1797, 2083, 2484, 3104]},
    {"community_size": 6, "number_of_communities": 6, "sample_community_ids": [1890, 2099, 239, 882, 2317]},
    {"community_size": 5, "number_of_communities": 8, "sample_community_ids": [1936, 635, 672, 996, 1177]},
    {"community_size": 4, "number_of_communities": 17, "sample_community_ids": [1865, 459, 320, 1666, 557]},
    {"community_size": 3, "number_of_communities": 25, "sample_community_ids": [5, 137, 718, 906, 1165]},
    {"community_size": 2, "number_of_communities": 111, "sample_community_ids": [106, 107, 133, 138, 193]},
    {"community_size": 1, "number_of_communities": 906, "sample_community_ids": [9, 15, 18, 28, 29]}
]

# 2. En büyük toplulukların varlık türü dağılımı
entity_distribution_data = [
    {
        "community_id": 493, "total_size": 329,
        "entity_distribution": [
            {"count": 47, "entity_type": "FOCUS_AREA_THEME"},
            {"count": 202, "entity_type": "SOLUTION_APPROACH"},
            {"count": 41, "entity_type": "PROBLEM_CHALLENGE"},
            {"count": 39, "entity_type": "STAKEHOLDER"}
        ]
    },
    {
        "community_id": 1491, "total_size": 296,
        "entity_distribution": [
            {"count": 32, "entity_type": "STAKEHOLDER"},
            {"count": 52, "entity_type": "PROBLEM_CHALLENGE"},
            {"count": 158, "entity_type": "SOLUTION_APPROACH"},
            {"count": 54, "entity_type": "FOCUS_AREA_THEME"}
        ]
    },
    {
        "community_id": 1541, "total_size": 228,
        "entity_distribution": [
            {"count": 29, "entity_type": "STAKEHOLDER"},
            {"count": 36, "entity_type": "FOCUS_AREA_THEME"},
            {"count": 120, "entity_type": "SOLUTION_APPROACH"},
            {"count": 43, "entity_type": "PROBLEM_CHALLENGE"}
        ]
    },
    {
        "community_id": 2345, "total_size": 153,
        "entity_distribution": [
            {"count": 21, "entity_type": "STAKEHOLDER"},
            {"count": 33, "entity_type": "PROBLEM_CHALLENGE"},
            {"count": 82, "entity_type": "SOLUTION_APPROACH"},
            {"count": 17, "entity_type": "FOCUS_AREA_THEME"}
        ]
    },
    {
        "community_id": 199, "total_size": 119,
        "entity_distribution": [
            {"count": 14, "entity_type": "STAKEHOLDER"},
            {"count": 31, "entity_type": "PROBLEM_CHALLENGE"},
            {"count": 67, "entity_type": "SOLUTION_APPROACH"},
            {"count": 7, "entity_type": "FOCUS_AREA_THEME"}
        ]
    },
    {
        "community_id": 2257, "total_size": 111,
        "entity_distribution": [
            {"count": 7, "entity_type": "STAKEHOLDER"},
            {"count": 18, "entity_type": "FOCUS_AREA_THEME"},
            {"count": 67, "entity_type": "SOLUTION_APPROACH"},
            {"count": 19, "entity_type": "PROBLEM_CHALLENGE"}
        ]
    }
]

# 3. Topluluk içi bağlantı yoğunluğu
density_data = [
    {"community_id": 2066, "community_size": 10, "internal_edges": 12, "density": 0.26666666666666666, "avg_degree_internal": 1},
    {"community_id": 765, "community_size": 10, "internal_edges": 11, "density": 0.24444444444444444, "avg_degree_internal": 1},
    {"community_id": 735, "community_size": 10, "internal_edges": 9, "density": 0.2, "avg_degree_internal": 0},
    {"community_id": 236, "community_size": 15, "internal_edges": 20, "density": 0.19047619047619047, "avg_degree_internal": 1},
    {"community_id": 974, "community_size": 11, "internal_edges": 10, "density": 0.18181818181818182, "avg_degree_internal": 0},
    {"community_id": 1863, "community_size": 10, "internal_edges": 8, "density": 0.17777777777777778, "avg_degree_internal": 0},
    {"community_id": 2574, "community_size": 19, "internal_edges": 24, "density": 0.14035087719298245, "avg_degree_internal": 1},
    {"community_id": 852, "community_size": 24, "internal_edges": 35, "density": 0.12681159420289856, "avg_degree_internal": 1},
    {"community_id": 2828, "community_size": 19, "internal_edges": 21, "density": 0.12280701754385964, "avg_degree_internal": 1},
    {"community_id": 1967, "community_size": 20, "internal_edges": 23, "density": 0.12105263157894737, "avg_degree_internal": 1}
]

# 1. Şekil 5.3.1: Topluluk Boyut Dağılımı
def create_community_size_distribution():
    # Sadece 10 ve üzeri topluluk boyutlarını göster
    df = pd.DataFrame(community_size_data)
    df_filtered = df[df['community_size'] >= 10].head(15)  # İlk 15 büyük topluluk
    
    plt.figure(figsize=(14, 8))
    
    # Topluluk sayısı ile ağırlıklandırılmış çubuk grafiği
    bars = plt.bar(range(len(df_filtered)), 
                   df_filtered['community_size'], 
                   color='#45B7D1', 
                   alpha=0.8,
                   edgecolor='white',
                   linewidth=1)
    
    # X ekseni etiketleri
    labels = [f"{size}\n({count} adet)" for size, count in 
              zip(df_filtered['community_size'], df_filtered['number_of_communities'])]
    plt.xticks(range(len(df_filtered)), labels, fontsize=9, rotation=45, ha='right')
    
    # Eksen formatı
    plt.xlabel('Topluluk Boyutu (Topluluk Sayısı)', fontsize=12, fontweight='bold')
    plt.ylabel('Topluluk Üye Sayısı', fontsize=12, fontweight='bold')
    plt.title('Topluluk Boyut Dağılımı (10+ Üyeli Topluluklar)', 
              fontsize=14, fontweight='bold', pad=20)
    
    # Grid
    plt.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Değer etiketleri
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'{int(height)}', ha='center', va='bottom', fontsize=9)
    
    # Y eksenini optimize et
    plt.ylim(0, max(df_filtered['community_size']) * 1.1)
    
    plt.tight_layout()
    plt.savefig('Şekil_5_3_1_Topluluk_Boyut_Dağılımı.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 2. Şekil 5.3.2: Topluluk İçi Varlık Türü Dağılımı
def create_entity_type_distribution():
    # Veriyi yığılmış çubuk grafik için hazırla
    communities = []
    stakeholder_counts = []
    problem_counts = []
    solution_counts = []
    focus_counts = []
    
    for comm_data in entity_distribution_data:
        communities.append(f"Topluluk {comm_data['community_id']}\n({comm_data['total_size']} üye)")
        
        # Her varlık türü sayısını al
        type_counts = {dist['entity_type']: dist['count'] for dist in comm_data['entity_distribution']}
        stakeholder_counts.append(type_counts.get('STAKEHOLDER', 0))
        problem_counts.append(type_counts.get('PROBLEM_CHALLENGE', 0))
        solution_counts.append(type_counts.get('SOLUTION_APPROACH', 0))
        focus_counts.append(type_counts.get('FOCUS_AREA_THEME', 0))
    
    plt.figure(figsize=(12, 8))
    
    # Yığılmış çubuk grafik
    width = 0.6
    x = np.arange(len(communities))
    
    # Her varlık türü için çubuklar
    p1 = plt.bar(x, stakeholder_counts, width, 
                label='Paydaş (STAKEHOLDER)', color=color_map['STAKEHOLDER'], alpha=0.8)
    p2 = plt.bar(x, problem_counts, width, bottom=stakeholder_counts,
                label='Problem/Zorluk (PROBLEM)', color=color_map['PROBLEM_CHALLENGE'], alpha=0.8)
    p3 = plt.bar(x, solution_counts, width, 
                bottom=np.array(stakeholder_counts) + np.array(problem_counts),
                label='Çözüm/Yaklaşım (SOLUTION)', color=color_map['SOLUTION_APPROACH'], alpha=0.8)
    p4 = plt.bar(x, focus_counts, width,
                bottom=np.array(stakeholder_counts) + np.array(problem_counts) + np.array(solution_counts),
                label='Odak Alanı/Tema (FOCUS)', color=color_map['FOCUS_AREA_THEME'], alpha=0.8)
    
    # Eksen formatı
    plt.xlabel('Topluluklar', fontsize=12, fontweight='bold')
    plt.ylabel('Varlık Sayısı', fontsize=12, fontweight='bold')
    plt.title('Topluluk İçi Varlık Türü Dağılımı (İlk 6 Büyük Topluluk)', 
              fontsize=14, fontweight='bold', pad=20)
    
    # X ekseni etiketleri
    plt.xticks(x, communities, fontsize=9, rotation=45, ha='right')
    
    # Legend ve grid
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
    plt.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Yüzde etiketleri ekle (her bölüm için)
    for i, (sh, pr, so, fo) in enumerate(zip(stakeholder_counts, problem_counts, solution_counts, focus_counts)):
        total = sh + pr + so + fo
        # SOLUTION_APPROACH yüzdesi (en büyük bölüm)
        solution_pct = (so / total) * 100
        plt.text(i, sh + pr + so/2, f'{solution_pct:.1f}%', 
                ha='center', va='center', fontweight='bold', color='white')
    
    plt.tight_layout()
    plt.savefig('Şekil_5_3_2_Topluluk_İçi_Varlık_Türü_Dağılımı.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 3. Şekil 5.3.3: Topluluk İçi Bağlantı Yoğunluğu
def create_internal_density_scatter():
    df = pd.DataFrame(density_data)
    
    plt.figure(figsize=(12, 8))
    
    # Scatter plot with size based on internal edges
    sizes = [edge_count * 30 for edge_count in df['internal_edges']]  # Scale for visibility
    colors = [size for size in df['community_size']]  # Color by community size
    
    scatter = plt.scatter(df['community_size'], df['density'], 
                         s=sizes, c=colors, alpha=0.7, 
                         cmap='viridis', edgecolors='white', linewidth=1)
    
    # Her nokta için topluluk ID'sini etiket olarak ekle
    for i, row in df.iterrows():
        plt.annotate(f"T{int(row['community_id'])}", 
                    (row['community_size'], row['density']),
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=9, fontweight='bold')
    
    # Eksen formatı
    plt.xlabel('Topluluk Boyutu', fontsize=12, fontweight='bold')
    plt.ylabel('İç Bağlantı Yoğunluğu (Density)', fontsize=12, fontweight='bold')
    plt.title('Topluluk İçi Bağlantı Yoğunluğu', 
              fontsize=14, fontweight='bold', pad=20)
    
    # Grid
    plt.grid(True, alpha=0.3, linestyle='--')
    
    # Colorbar for community size
    cbar = plt.colorbar(scatter)
    cbar.set_label('Topluluk Boyutu', fontsize=11)
    
    # Size legend
    legend_elements = []
    for size in [5, 10, 20, 35]:
        legend_elements.append(plt.scatter([], [], s=size*30, c='gray', alpha=0.7,
                                         label=f'{size} iç bağlantı'))
    plt.legend(handles=legend_elements, title='İç Bağlantı Sayısı', 
              loc='upper right', fontsize=9)
    
    # Y eksenini optimize et
    plt.ylim(0, max(df['density']) * 1.05)
    
    plt.tight_layout()
    plt.savefig('Şekil_5_3_3_Topluluk_İçi_Bağlantı_Yoğunluğu.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# Tüm grafikları oluştur
if __name__ == "__main__":
    print("Topluluk analizi görselleştirmeleri oluşturuluyor...")
    
    create_community_size_distribution()
    print("✓ Şekil 5.3.1 oluşturuldu")
    
    create_entity_type_distribution()
    print("✓ Şekil 5.3.2 oluşturuldu")
    
    create_internal_density_scatter()
    print("✓ Şekil 5.3.3 oluşturuldu")
    
    print("Tüm görselleştirmeler tamamlandı!")
    print("\nNot: Topluluklar arası bağlantı ağı (Şekil 5.3.4) için Neo4j Browser kullanabilirsiniz.")
```
