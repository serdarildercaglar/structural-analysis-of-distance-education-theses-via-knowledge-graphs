```python
import matplotlib.pyplot as plt
import numpy as np

# Veriler
communities = ['1927', '1201', '2738', '385', '505']
stakeholders = [17, 28, 15, 28, 6]
problems = [27, 82, 149, 62, 26]
solutions = [83, 202, 130, 95, 77]
focus_areas = [119, 99, 23, 13, 15]

# Toplamlar hesapla
totals = [sum(x) for x in zip(stakeholders, problems, solutions, focus_areas)]

# Yzdeleri hesapla
stakeholders_perc = [s/t*100 for s, t in zip(stakeholders, totals)]
problems_perc = [p/t*100 for p, t in zip(problems, totals)]
solutions_perc = [s/t*100 for s, t in zip(solutions, totals)]
focus_areas_perc = [f/t*100 for f, t in zip(focus_areas, totals)]

# Grafii olutur
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Sol taraf: Mutlak deerler
width = 0.6
ind = np.arange(len(communities))

p1 = ax1.bar(ind, stakeholders, width, color='#ff9999', label='STAKEHOLDER')
p2 = ax1.bar(ind, problems, width, bottom=stakeholders, color='#66b3ff', label='PROBLEM_CHALLENGE')
p3 = ax1.bar(ind, solutions, width, bottom=[sum(x) for x in zip(stakeholders, problems)], 
            color='#99ff99', label='SOLUTION_APPROACH')
p4 = ax1.bar(ind, focus_areas, width, bottom=[sum(x) for x in zip(stakeholders, problems, solutions)], 
            color='#ffcc99', label='FOCUS_AREA_THEME')

ax1.set_ylabel('Dm Says', fontsize=12)
ax1.set_title('A. Topluluklarn Varlk Trlerine Gre Dalm (Mutlak)', fontsize=14)
ax1.set_xticks(ind)
ax1.set_xticklabels(['Topluluk '+c for c in communities])
ax1.legend(loc='upper right')

# ubuklarn zerine toplam deerleri ekle
for i, total in enumerate(totals):
    ax1.text(i, total+5, str(total), ha='center', fontsize=10, fontweight='bold')

# Sa taraf: Yzdelik deerler
p1 = ax2.bar(ind, stakeholders_perc, width, color='#ff9999', label='STAKEHOLDER')
p2 = ax2.bar(ind, problems_perc, width, bottom=stakeholders_perc, color='#66b3ff', label='PROBLEM_CHALLENGE')
p3 = ax2.bar(ind, solutions_perc, width, bottom=[sum(x) for x in zip(stakeholders_perc, problems_perc)], 
            color='#99ff99', label='SOLUTION_APPROACH')
p4 = ax2.bar(ind, focus_areas_perc, width, bottom=[sum(x) for x in zip(stakeholders_perc, problems_perc, solutions_perc)], 
            color='#ffcc99', label='FOCUS_AREA_THEME')

ax2.set_ylabel('Dm Oran (%)', fontsize=12)
ax2.set_title('B. Topluluklarn Varlk Trlerine Gre Dalm (Yzde)', fontsize=14)
ax2.set_xticks(ind)
ax2.set_xticklabels(['Topluluk '+c for c in communities])
ax2.legend(loc='upper right')

plt.suptitle('ekil 7: nemli Topluluklarn Varlk Trlerine Gre Dalm', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('community_entity_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
